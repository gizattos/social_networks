```python
import time
import tweepy
import pandas as pd
```

```python
##autenticando
auth = tweepy.OAuthHandler("API key", "API secret key")
auth.set_access_token("Access Token", "Secret Token")
api = tweepy.API(auth, wait_on_rate_limit=True)
```

```python
dict = {}
i=0
for user in tweepy.Cursor(api.friends, screen_name="gilmarmendes").items():
    dict[i] = {'user':user.screen_name,'nome':user.name,"link":"https://twitter.com/"+user.screen_name,"seguidores":user.followers_count}
    i=i+1
df = pd.DataFrame.from_dict(dict, "index")
```

```python
twitado=[]
twets = api.user_timeline(screen_name = 'alexandre', count = 100, include_rts = False)
for postagem in twets:
    twitado.append({'user': postagem.user.screen_name,'twittado': postagem.text,'curtidas': postagem.favorite_count,'rt': postagem.retweet_count,'data':postagem.created_at.strftime('%d/%m/%Y')})
```
